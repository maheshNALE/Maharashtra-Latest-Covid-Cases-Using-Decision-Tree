# **Maharashtra Latest Covid Cases Using Decision Tree**

# **Get Understanding about Data set**

**District-wise Covid-19 data of Maharashtra,a state in India as on January 28, 2022.**

**The data include number of positive cases, active cases, recovered, deceased cases, recovery rate and fatality rate.**

1.Districts - Name of districts in Maharashtra, India

2.Positive Cases - Number of positive cases

3.Active Cases - Number of active cases

4.Recovered - Number of recovered cases

5.Deceased - Number of deaths

6.Recovery Rate (%) - Ratio of number of recovered cases to positive cases

7.Fatality Rate (%) - Ratio of number of deaths to positive cases

# **Define y (dependent or label or target variable) and X (independent or features or attribute variable)**

# **Get X Variables Standardized**

Standardization of datasets is a common requirement for many machine learning estimators implemented in scikit-learn;they might behave badly if the individual features do not more or less look like standard normally distributed data:Gaussian with zero mean and unit variance.

Next approach is go for MinMax

```python
X_std=X[['Recovery Rate (%)','Fatality Rate (%)']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X_std=ss.fit_transform(X_std)
```

```python
X_std
```

```python
X[['Recovery Rate (%)','Fatality Rate (%)']]=pd.DataFrame(X_std,columns=['Recovery Rate (%)','Fatality Rate (%)'])
```

```python
X
```

# **Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Get Model Train**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=3,random_state=2529)
```

```python
dt.fit(X_train,y_train)
```

# **Get Model Prediction**

```python
y_pred=dt.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# **Get Probability of Each Predicted Class**

```python
dt.predict_proba(X_test)
```

# **Get Model Evaluation**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
